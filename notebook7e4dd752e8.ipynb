```python
augmentation_pipeline = A.Compose([
    A.Perspective(scale=(0.05, 0.15), p=0.5),
    A.Affine(rotate=(-30, 30), shear=(-20, 20), scale=(0.8, 1.2), p=0.5),
    A.RandomScale(scale_limit=0.4, p=0.5),
    A.RandomCrop(width=640, height=640, p=0.5),
    #A.CoarseDropout(max_holes=12, max_height=60, max_width=60, fill_value=0, p=0.5),
    A.CoarseDropout( num_holes_range=(3, 12),  hole_height_range=(10, 60),    hole_width_range=(10, 60), fill="random_uniform",  p=1.0),
    A.MotionBlur(blur_limit=9, p=0.5),
    A.RandomBrightnessContrast(brightness_limit=0.4, contrast_limit=0.4, p=0.5),
    A.GaussianBlur(blur_limit=(3, 7), p=0.5),
    A.GaussNoise(std_range=(0.1,0.5), p=0.5)
], bbox_params=A.BboxParams(format='yolo', label_fields=['class_labels']))
```

```python
import cv2
from tqdm import tqdm
from glob import glob
```

```python
image_dir = "/kaggle/input/iruvd-dataset-for-automatic-vehicle-detection/JU_yolov5/train/images"
label_dir = "/kaggle/input/iruvd-dataset-for-automatic-vehicle-detection/JU_yolov5/train/labels"
output_image_dir = "/kaggle/working/augmented_images/train/images"
output_label_dir = "/kaggle/working/augmented_images/train/labels"

os.makedirs(output_image_dir, exist_ok=True)
os.makedirs(output_label_dir, exist_ok=True)
```

```python
image_paths = glob(os.path.join(image_dir, "*.jpg"))

for img_path in tqdm(image_paths, desc="Augmenting Images", unit="img"):
    # Load image
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    height, width, _ = img.shape
    
    # Get corresponding label file
    label_path = os.path.join(label_dir, os.path.basename(img_path).replace(".jpg", ".txt"))
    
    if not os.path.exists(label_path):
        continue  # Skip images with no labels
    
    # Read bounding box labels
    with open(label_path, "r") as f:
        lines = f.readlines()
    
    bboxes = []
    class_labels = []
    
    for line in lines:
        data = line.strip().split()
        class_id = int(data[0])
        x_center, y_center, bbox_width, bbox_height = map(float, data[1:])
        
        # Convert bbox from YOLO format to normalized format
        bboxes.append([x_center, y_center, bbox_width, bbox_height])
        class_labels.append(class_id)
    
    # Apply augmentation
    augmented = augmentation_pipeline(image=img, bboxes=bboxes, class_labels=class_labels)

    # Save augmented image
    output_img_path = os.path.join(output_image_dir, os.path.basename(img_path))
    cv2.imwrite(output_img_path, cv2.cvtColor(augmented["image"], cv2.COLOR_RGB2BGR))
    
    # Save updated bounding boxes
    output_label_path = os.path.join(output_label_dir, os.path.basename(label_path))
    with open(output_label_path, "w") as f:
        for bbox, class_id in zip(augmented["bboxes"], augmented["class_labels"]):
            f.write(f"{class_id} {bbox[0]} {bbox[1]} {bbox[2]} {bbox[3]}\n")

print("Augmentation complete. Augmented images and labels saved.")
```

```python
import os
import json
import glob
import tqdm

# Define paths
IMG_DIR = "/kaggle/input/iruvd-dataset-for-automatic-vehicle-detection/JU_yolov5/valid/images"
LABEL_DIR = "/kaggle/input/iruvd-dataset-for-automatic-vehicle-detection/JU_yolov5/valid/labels"
ANNOT_FILE = "/kaggle/working/valid-annotations.json"

# Define class names
class_names = ['trak', 'cyclist', 'bike', 'tempo', 'car', 'zeep', 'toto', 'e-rickshaw', 
               'auto-rickshaw', 'bus', 'van', 'cycle-rickshaw', 'person', 'taxi']

# Initialize COCO format dictionary
coco_dict = {
    "images": [],
    "annotations": [],
    "categories": [{"id": i, "name": name} for i, name in enumerate(class_names)]
}

# Get all image files
image_files = glob.glob(os.path.join(IMG_DIR, "*.jpg"))

annotation_id = 0
for img_id, img_path in enumerate(tqdm.tqdm(image_files, desc="Processing Images")):
    img_name = os.path.basename(img_path)
    label_path = os.path.join(LABEL_DIR, img_name.replace(".jpg", ".txt"))

    if not os.path.exists(label_path):
        continue

    # Read image dimensions
    img = cv2.imread(img_path)
    height, width, _ = img.shape

    # Add image info
    coco_dict["images"].append({
        "id": img_id,
        "file_name": img_name,
        "width": width,
        "height": height
    })

    # Read YOLO labels
    with open(label_path, "r") as f:
        for line in f.readlines():
            label_data = line.strip().split()
            class_id = int(float(label_data[0]))  # Convert float to int safely
            x_center, y_center, bbox_width, bbox_height = map(float, label_data[1:])
            x_center, y_center, bbox_width, bbox_height = map(float, label_data[1:])

            # Convert YOLO to COCO bbox format (x_min, y_min, width, height)
            x_min = (x_center - bbox_width / 2) * width
            y_min = (y_center - bbox_height / 2) * height
            bbox_width *= width
            bbox_height *= height

            coco_dict["annotations"].append({
                "id": annotation_id,
                "image_id": img_id,
                "category_id": class_id,
                "bbox": [x_min, y_min, bbox_width, bbox_height],
                "area": bbox_width * bbox_height,
                "iscrowd": 0
            })
            annotation_id += 1

# Save to JSON
with open(ANNOT_FILE, "w") as f:
    json.dump(coco_dict, f, indent=4)

print(f"Annotations saved to {ANNOT_FILE}")
```

```python
# Training data
train_dataset = load_dataset(
    "json",  # Changed from "coco"
    data_files=train_labels_path,
    field="images",
    data_dir=train_images_dir
)

# Validation data
val_dataset = load_dataset(
    "json",
    data_files=valid_labels_path , 
    field="images", 
    data_dir=valid_images_dir
)
```

```python
with open(train_labels_path, "r") as f:
    annotations = json.load(f)
# Extract category names from the "categories" key in your COCO file
label_names = [cat["name"] for cat in annotations["categories"]]
num_labels = len(label_names)
print(f"Number of labels: {num_labels}")
print(f"Labels: {label_names}")
```

```python
image_processor = AutoImageProcessor.from_pretrained("facebook/detr-resnet-50")
model = DetrForObjectDetection.from_pretrained(
    "facebook/detr-resnet-50",
    num_labels=num_labels,
    ignore_mismatched_sizes=True)  # Necessary if the head size in the pretrained model differs
```

```python
def preprocess(example, image_dir):
    # Use the "file_name" field directly (it is inside the "images" object)
    file_name = example["file_name"]
    
    # Build the full image path (if file_name is not absolute)
    image_path = file_name if os.path.isabs(file_name) else os.path.join(image_dir, file_name)
    
    # Open and convert the image to RGB
    try:
        image = Image.open(image_path).convert("RGB")
    except Exception as e:
        raise ValueError(f"Error opening image at {image_path}: {e}")
    
    # Automatically resize and normalize the image for DETR
    inputs = image_processor(images=image, return_tensors="pt")
    
    # Remove the extra batch dimension and add as a new field "pixel_values"
    example["pixel_values"] = inputs["pixel_values"][0]
    return example
```

```python
cache_dir = "/kaggle/temp/cache"
os.makedirs(cache_dir, exist_ok=True)
```

```python
train_dataset = train_dataset["train"].map(
    lambda x: preprocess(x, train_images_dir),
    batched=False,
    writer_batch_size=1,
    load_from_cache_file=False,
    cache_file_name=os.path.join(cache_dir, "train_cache.arrow")
)

val_dataset = val_dataset["train"].map(
    lambda x: preprocess(x, valid_images_dir),
    batched=False,
    writer_batch_size=1,
    load_from_cache_file=False,
    cache_file_name=os.path.join(cache_dir, "val_cache.arrow")
)
```

```python
from transformers import TrainerCallback, TrainingArguments, Trainer, default_data_collator
```

```python
training_args = TrainingArguments(
    output_dir="./codetr-output",            # directory to save checkpoints and logs
    run_name="codetr_run_10epochs",           # custom name for the current run
    num_train_epochs=3,                      # train for 10 epochs
    per_device_train_batch_size=2,            # use a modest batch size
    gradient_accumulation_steps=2,            # effectively doubles batch size
    learning_rate=1e-4,
    evaluation_strategy="epoch",
    save_strategy="epoch",
    logging_steps=50,
    fp16=True,                              # use mixed precision if your GPU supports it
    logging_dir="./logs",
    remove_unused_columns=False,
    disable_tqdm=False
)
```

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    data_collator=default_data_collator
)
```

```python
trainer.train()
```

```python
classes=['trak', 'cyclist', 'bike', 'tempo', 'car', 'zeep', 'toto', 'e-rickshaw', 'auto-rickshaw', 'bus', 'van', 'cycle-rickshaw', 'person', 'taxi']
```

```python
import os
import json
from glob import glob

# Define paths
images_dir = "/kaggle/input/iruvd-dataset-for-automatic-vehicle-detection/JU_yolov5/valid/images"
labels_dir = "/kaggle/input/iruvd-dataset-for-automatic-vehicle-detection/JU_yolov5/valid/labels"
output_json = "/kaggle/working/valid_coco.json"

# Find images
image_files = glob(os.path.join(images_dir, "*.*"))  
valid_extensions = ('.jpg', '.jpeg', '.png')
image_files = [f for f in image_files if f.lower().endswith(valid_extensions)]
print(f"🔍 Total images found: {len(image_files)}")

coco_data = {
    "images": [],
    "annotations": [],
    "categories": [{"id": i, "name": name} for i, name in enumerate(classes)]
}

image_id, annotation_id, missing_labels, skipped_images = 0, 0, 0, 0

for img_path in image_files:
    img_filename = os.path.basename(img_path)
    label_path = os.path.join(labels_dir, img_filename.rsplit(".", 1)[0] + ".txt")  

    if not os.path.exists(label_path):
        missing_labels += 1
        continue  

    with open(label_path, "r") as f:
        lines = f.readlines()
        if len(lines) == 0:
            print(f"⚠️ Empty label file: {label_path}")  
            skipped_images += 1
            continue  

        # Add image metadata
        coco_data["images"].append({
            "id": image_id,
            "file_name": img_filename,
            "width": 1920, "height": 1080  
        })

        valid_annotations = 0
        for line in lines:
            parts = line.strip().split()
            if len(parts) != 5:
                print(f"❌ Invalid label format in: {label_path}")  
                skipped_images += 1
                continue  

            cls_id, x_center, y_center, w, h = map(float, parts)
            coco_data["annotations"].append({
                "id": annotation_id,
                "image_id": image_id,
                "category_id": int(cls_id),
                "bbox": [
                    (x_center - w / 2) * 1920,  
                    (y_center - h / 2) * 1080,
                    w * 1920,
                    h * 1080
                ],
                "area": w * h * 1920 * 1080,
                "iscrowd": 0
            })
            annotation_id += 1
            valid_annotations += 1  

        if valid_annotations == 0:
            print(f"🚨 No valid objects in: {label_path}")  
            skipped_images += 1
            continue  

        image_id += 1

# Save COCO JSON
with open(output_json, "w") as f:
    json.dump(coco_data, f, indent=4)

print(f"\n✅ COCO annotations saved: {output_json}")
print(f"📌 Total images processed: {image_id} / {len(image_files)}")
print(f"⚠️ Missing labels: {missing_labels}")
print(f"🚨 Skipped images: {skipped_images}")
print(f"📌 Total annotations: {annotation_id}")
```
